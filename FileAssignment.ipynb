```python
#QUESTION NO.1
source = input("Enter source file name: ")
destination = input("Enter destination file name: ")
file1 = open(source, "r")
content = file1.read()
upper_content = content.upper()
file2 = open(destination, "w")
file2.write(upper_content)
file1.close()
file2.close()
print("File copied with uppercase content successfully.")
```

```python
#QUESTION NO.2
source = input("Enter source python file: ")
destination = input("Enter destination file: ")

file1 = open(source, "r")
file2 = open(destination, "w")

for line in file1:
    if not line.strip().startswith("#"):
        file2.write(line)

file1.close()
file2.close()

print("\nContent of Source File:\n")

file1 = open(source, "r")
print(file1.read())
file1.close()

print("\nContent of Destination File (without comments):\n")

file2 = open(destination, "r")
print(file2.read())
file2.close()
```
