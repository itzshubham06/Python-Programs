```python
def is_perfect_number(number):
  if number <= 0:
    return False # Perfect numbers are positive integers

  sum_divisors = 0
  for i in range(1, number):
    if number % i == 0:
      sum_divisors += i

  return sum_divisors == number
print(f"Is 6 a perfect number? {is_perfect_number(6)}")    # Expected: True
print(f"Is 28 a perfect number? {is_perfect_number(28)}")  # Expected: True
print(f"Is 10 a perfect number? {is_perfect_number(10)}")  # Expected: False
print(f"Is 1 a perfect number? {is_perfect_number(1)}")    # Expected: False
print(f"Is 0 a perfect number? {is_perfect_number(0)}")    # Expected: False
```

```python
import re

def is_palindrome(s):

  cleaned_s = re.sub(r'[^a-z0-9]', '', s.lower())
  return cleaned_s == cleaned_s[::-1]

print(f"'madam' is a palindrome: {is_palindrome('madam')}")
print(f"'nurses run' is a palindrome: {is_palindrome('nurses run')}")
print(f"'A man, a plan, a canal: Panama' is a palindrome: {is_palindrome('A man, a plan, a canal: Panama')}")
print(f"'hello' is a palindrome: {is_palindrome('hello')}")
print(f"'Racecar' is a palindrome: {is_palindrome('Racecar')}")
```

```python
import string

def is_pangram(s):
  s_lower = s.lower()
  present_letters = set()
  for char in s_lower:
    if 'a' <= char <= 'z':
      present_letters.add(char)
  all_alphabet_letters = set(string.ascii_lowercase)
  return all_alphabet_letters.issubset(present_letters)
example_string = 'The quick brown fox jumps over the lazy dog'
is_pangram_result = is_pangram(example_string)

print(f"'{example_string}' is a pangram: {is_pangram_result}")
print(f"'Hello World' is a pangram: {is_pangram('Hello World')}")
print(f"'abcdefghijklmnopqrstuvwxyz' is a pangram: {is_pangram('abcdefghijklmnopqrstuvwxyz')}")
print(f"'Pack my box with five dozen liquor jugs' is a pangram: {is_pangram('Pack my box with five dozen liquor jugs')}")
```

```python
def sort_hyphenated_words(words_string):
  words = words_string.split('-')
  words.sort()
  sorted_string = '-'.join(words)
  print(sorted_string)
example_input = 'green-red-yellow-black-white'
print(f"Original string: '{example_input}'")
print("Sorted string:")
sort_hyphenated_words(example_input)
```

```python
def make_bold(func):
  def wrapper(*args, **kwargs):
    return "<b>" + func(*args, **kwargs) + "</b>"
  return wrapper

def make_italic(func):
  def wrapper(*args, **kwargs):
    return "<i>" + func(*args, **kwargs) + "</i>"
  return wrapper

def make_underline(func):
  def wrapper(*args, **kwargs):
    return "<u>" + func(*args, **kwargs) + "</u>"
  return wrapper
@make_bold
@make_italic
@make_underline
def hello_world():
  return "Hello World"
print(hello_world())
```
