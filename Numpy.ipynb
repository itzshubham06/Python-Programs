```python
s1="SBJAIN college 54678"
print(s1)
cspace=0
cdigit=0
calphabet=0
for i in s1:
  if i.isspace():
    cspace+=1
  if i.isdigit():
    cdigit+=1
  if i.isalnum():
    calphabet+=1
print("String =",s1)
print("Space =",cspace)
print("Digit =",cdigit)
print("Alphabet =",calphabet)
print(s1.count("s"))
```
